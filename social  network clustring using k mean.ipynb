# Select Features for K-Means

### Select only the 'Age' and 'EstimatedSalary' columns from the dataset
### These columns are stored in a new variable X
### X contains the features (input variables) that will be used for clustering

```python
X = df[['Age','EstimatedSalary']]

print(X.head())
```

# Standard Scaler

#### Create an object of StandardScaler
#### StandardScaler is used to normalize the data
#### It converts values to a common scale with:
#### Mean = 0
#### Standard Deviation = 1
#### Fit the scaler on the data and transform it
#### fit() calculates the mean and standard deviation of each column
#### transform() scales the values using those statistics
#### fit_transform() performs both steps together
#### Display the first 5 rows of the scaled data

```python
scaler = StandardScaler()


X_scaled = scaler.fit_transform(X)


print(X_scaled[:5])
```

# Elbow Method

#### Create an empty list to store WCSS values
#### WCSS = Within-Cluster Sum of Squares
#### It measures how close the data points are to their cluster center
#### Lower WCSS means better clustering
#### Create a K-Means model with i clusters
#### Train the K-Means model on the scaled data
#### Store the WCSS (inertia) value in the list
#### inertia_ gives the total distance of points from their cluster centroids

```python
wcss = []
for i in range(1,11):  # Try different numbers of clusters from 1 to 10
    kmeans = KMeans(
        n_clusters=i,       # Number of clusters
        init='k-means++',   # Smart method for choosing initial centroids
        random_state=42,    # Ensures the same result every time the code runs
        n_init=10            # Run K-Means 10 times and choose the best result
    ) 
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

# Elbow Graph

#### Create a new figure for the grap
#### figsize=(8,5) sets the width to 8 inches and height to 5 inches
#### Plot the Elbow Method graph
#### range(1,11) represents the number of clusters (K) from 1 to 10
#### wcss contains the corresponding WCSS values
#### marker='o' displays a circle at each data point

```python
                               
plt.figure(figsize=(8,5))
                                    
plt.plot(range(1,11),wcss,marker='o')

plt.title("Elbow Method")         # Add a title to the graph
plt.xlabel("Number of Clusters") # Label the x-axis
                               # Shows the number of clusters being tested
plt.ylabel("WCSS")              # Label the y-axis
                              # Shows the WCSS value for each cluster coun
plt.grid(True)
plt.show()
```

# Apply K-Means

#### Create a K-Means model with 3 clusters
#### n_clusters=3 means the data will be divided into 3 groups
#### init='k-means++' selects initial centroids in a smart way
#### random_state=42 ensures the same results every time the code runs
#### n_init=10 runs K-Means 10 times and chooses the best clustering result
#### Apply K-Means clustering to the scaled dataset
#### fit_predict() performs two tasks:
#### 1. Trains the K-Means model
#### 2. Assigns a cluster label to each data point
#### The cluster labels are stored in the variable 'clusters'

```python
kmeans = KMeans(
    n_clusters=3,
    init='k-means++',
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
```

# Cluster Result

#### Print a heading to make the output easier to understand
#### This message indicates that the clustering results will be displayed
#### Select the columns Age, EstimatedSalary, and Cluster from the DataFrame
#### Age = person's age
#### EstimatedSalary = person's estimated salary
#### Cluster = group assigned by the K-Means algorithm
#### .head(20) displays the first 20 rows of the selected columns
#### This helps us quickly check how the data has been grouped into clusters


```python
print("Cluster Result")

print(df[['Age','EstimatedSalary','Cluster']].head(20))
```

# Cluster Statistics

#### df.groupby('Cluster') → Groups all records based on their cluster number.
#### [['Age', 'EstimatedSalary']] → Selects the Age and EstimatedSalary columns.
#### . mean() → Calculates the average value of these columns for each cluster.
####  print() → Displays the cluster summary.

```python
print(df.groupby('Cluster')[['Age','EstimatedSalary']].mean())
```

# K-Means Visualization

#### Create a new figure for the clustering graph
#### figsize=(10,6) sets the graph size (width=10, height=6)
#### Plot all data points that belong to Cluster 0
#### X_scaled[clusters==0,0] -> X-axis values (Age)
#### X_scaled[clusters==0,1] -> Y-axis values (Estimated Salary)
#### s=80 sets the size of each point
#### label is used in the graph legend
#### Plot the centroids (center points) of each cluster
#### cluster_centers_ contains the coordinates of the cluster centers
#### [:,0] -> X coordinates of centroids
#### [:,1] -> Y coordinates of centroids
#### s=300 makes centroid points larger
#### marker='X' displays centroids as X symbols
#### color='black' makes centroids black
#### Display the legend showing cluster names and centroids

```python
plt.figure(figsize=(10,6))

plt.scatter(
    X_scaled[clusters==0,0],
    X_scaled[clusters==0,1],
    s=80,
    label='Cluster 1'
)

plt.scatter(
    X_scaled[clusters==1,0],
    X_scaled[clusters==1,1],
    s=80,
    label='Cluster 2'
)

plt.scatter(
    X_scaled[clusters==2,0],
    X_scaled[clusters==2,1],
    s=80,
    label='Cluster 3'
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=300,
    marker='X',
    color='black',
    label='Centroids'
)

plt.title("K-Means Clustering Result")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()

plt.show()
```
